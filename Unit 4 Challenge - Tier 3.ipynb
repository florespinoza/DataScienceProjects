#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices  = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices , sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
# Rename the columns names to the first row date values using the iloc[] method:
properties_transposed.columns = properties_transposed.iloc[0]
# Check the head of the transposed DataFrame to see the the new column names:
properties_transposed.head()
```

```python
# Delete the first row transposed DataFrame: 
properties_transposed = properties_transposed.drop(0)
# Check the head of the transposed DataFrame to see the the deletion of the date values row:
properties_transposed.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Define a dictionary to rename the 'Unnamed: 0' and 'NaN' columns
colsrename = {'Unnamed: 0':'London_Borough', pd.NaT: 'ID'}
properties_transposed = properties_transposed.rename(columns = colsrename)

# Check the head of the transposed DataFrame to see the the deletion of the date row values:
properties_transposed.head()
```

```python
# The dimensions (number of rows, number of columns) of the cleaned data are:
print('(number of London boroughs, number of dates) =',properties_transposed.shape)
```

```python
# Print a view of the column names  
properties_transposed.columns
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# Use melt to list all columns data date values into one column and call the melted DataFrame df
properties_transformed = pd.melt(properties_transposed, id_vars= ['London_Borough', 'ID'])
properties_transformed.head()
```

```python
# Print the new dimensions (number of rows, number of columns) of the transformed DataFrame
properties_transformed.shape
```

```python
# Rename the '0' column to 'Month', and the 'value' column to 'Average_price'.
# First define a dictionary with the new column names.
# Then use the rename() method, and reassign clean_properties with the result:
cnames = {0: 'Month', 'value': 'Average_price'}
properties_transformed = properties_transformed.rename(columns = cnames)
properties_transformed.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Check the types of the transformed DataFrame variables
properties_transformed.dtypes
```

```python
# Change the Average_price column to a numeric type. Its values have to be floatting point numbers. 
# Use the to_numeric() method on pandas pd
properties_transformed['Average_price'] = pd.to_numeric(properties_transformed['Average_price'])
# Check the new type of the average price variable
properties_transformed.dtypes
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# print out all names listed as London borough, check if there are any corrupt bourough names in our data:
borough_names=properties_transformed['London_Borough'].unique()
borough_names
```

```python
# Gather and print the correct list containing London borough names including the 33 "city of London" borough:
london_boroughs_names = np.array(['City of London', 'Barking & Dagenham', 'Barnet', 'Bexley',
       'Brent', 'Bromley', 'Camden', 'Croydon', 'Ealing', 'Enfield',
       'Greenwich', 'Hackney', 'Hammersmith & Fulham', 'Haringey',
       'Harrow', 'Havering', 'Hillingdon', 'Hounslow', 'Islington',
       'Kensington & Chelsea', 'Kingston upon Thames', 'Lambeth',
       'Lewisham', 'Merton', 'Newham', 'Redbridge',
       'Richmond upon Thames', 'Southwark', 'Sutton', 'Tower Hamlets',
       'Waltham Forest', 'Wandsworth', 'Westminster'])
london_boroughs_names
```

```python
# Identify all rows from DataFrame df that are included in the london_boroughs np array
london_boroughs = properties_transformed[properties_transformed['London_Borough'].isin(london_boroughs_names)]
# Print the size of the cleaned Dataframe after removing all corrupt London borough names
london_boroughs.shape
```

```python
# Check the head of the cleaned DataFrame after removing all corrupt London borough names:
london_boroughs.head()
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
# Define a dictionary with the name of the selected borough:
borough ={'name': 'Barnet'}
# Define the variable borough_prices to extract the result of filtering df on the selected borough average price data:
borough_prices = london_boroughs[london_boroughs['London_Borough'] == borough['name']]

# Make a variable called ax. Assign it the result of calling the plot() method, and plugging in the following values as parameters:
# kind ='line', x = 'Month', y='Average_price'
ax = borough_prices.plot(kind ='line', x = 'Month', y='Average_price', title='%s Borough \'s Average Housing Prices'%borough['name'])

# Finally, call the set_ylabel() method on ax, and set that label to the string: 'Price'. 
ax.set_ylabel('Price')
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column. 

To this end, you *could* apply a ***lambda function***. Your logic could work as follows:
1. look through the `Month` column
2. extract the year from each individual value in that column 
3. store that corresponding year as separate column. 

Whether you go ahead with this is up to you. Just so long as you answer our initial brief: which boroughs of London have seen the greatest house price increase, on average, over the past two decades? 

```python
london_boroughs['Year'] = london_boroughs['Month'].apply(lambda t: t.year)
# Check the head of the cleaned DataFrame after removing all corrupt London borough names:
london_boroughs.head()
```

```python
# Compute the mean house price for each year grouping the dataFreme by the London_Borough and Year columns.
# Define the DataFrame df and asign it the result of calling the groupby() method on df 
# with the parameters by=['Borough', 'Year'], and apply the mean() to it.

df = london_boroughs.groupby(by = ['London_Borough', 'Year']).mean()
# Check the head of the grouped DataFrame showing 30 rows:
df.head(30)
```

```python
# The grouped DataFrame contains row blocks of boroughs with years and the average rent price for each year
# Print the size of the grouped Dataframe after grouping by London_Borough and Year columns
df.shape
```

```python
# As we can notice from the above results, we have reduced our data size significantly: 
print('Note: By grouping our data by year, we obtained a significant data reduction from %s to %s rows.' %(london_boroughs.shape[0], df.shape[0]))
```

```python
# Reset the index of the grouped DataFrame df_grouped:
df = df.reset_index()
# Check the head of the grouped DataFrame after resetting the index:
df.head()
```

```python
# Check the tail of the grouped DataFrame after resetting the index:
df.tail()
```

```python
# Check the info of the grouped DataFrame to see if the data contains any null values:
df.info()
```

```python
# To visualize the yearly data, we selecte a particular London Borough and make a line plot of Year against Average Price.
# Define a dictionary with the name of the selected borough:
borough ={'name': 'Westminster'}
# Define the variable borough_prices to extract the result of filtering df on the selected borough average price data:
borough_yearly_prices = df[df['London_Borough'] == borough['name']]
# Define the variable ax with plotting parameters
ax = borough_yearly_prices.plot(kind ='line', x = 'Year', y='Average_price', title='%s Borough \'s Yearly Average Rent Prices'%borough['name'])
ax.set_ylabel('Price')
```

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:

```python
# Subset the data for year 1998
year_1998 = df[df['Year']==1998]
year_1998.head()
```

```python
# Subset the data for year 2018
year_2018 = df[df['Year']==2018]
year_2018.head()
```

```python
# Merge the the year_1998 and year_2018 DataFrames into a DataFrame named df_decade
df_decade = pd.merge(year_1998, year_2018, on='London_Borough', suffixes=['_1998', '_2018'])
df_decade.head()
```

```python
# Delete the Year_1998 and Year_2018 columns 
df_decade.drop(['Year_1998', 'Year_2018'], axis=1, inplace=True)
df_decade.head()
```

```python
# Visualize the London Borough's average prices for years 1998 and 2018 together 
ax = df_decade[['Average_price_1998', 'Average_price_2018']].plot(kind='bar', title='London Boroughs Average Rent Prices')
ax.set_ylabel('Price')
# List London Boroughs on the x-axis
ax.set_xticklabels(df_decade.London_Borough)
```

```python
df_decade['Ratio_2018_1998'] = df_decade['Average_price_2018']/df_decade['Average_price_1998']
df_decade.head()
```

```python
# Visualize the London Borough's average prices for years 1998 and 2018 together 
ax = df_decade['Ratio_2018_1998'].plot(kind='bar', title='London Boroughs Average Housing Price 2018/1998 Ratios')
ax.set_ylabel('Price')
# List London Boroughs on the x-axis
ax.set_xticklabels(df_decade.London_Borough)
```

```python
# Sort the DataFrame by the ratio values to identify the London Boroughs with the highest Housing increase prices
df_sorted = df_decade.sort_values('Ratio_2018_1998', ascending = False)
df_sorted.head()
```

```python
# Visualize the London Borough's average prices for years 1998 and 2018 after sorting the ration values in descending order 
ax = df_sorted['Ratio_2018_1998'].plot(kind='bar', title='London Boroughs Average Rent Price 2018/1998 Ratios')
ax.set_ylabel('Price')
# List London Boroughs on the x-axis
ax.set_xticklabels(df_sorted.London_Borough)
```

```python
london_bourough_highest_increase = df_sorted.iloc[:15]
london_bourough_highest_increase
```
